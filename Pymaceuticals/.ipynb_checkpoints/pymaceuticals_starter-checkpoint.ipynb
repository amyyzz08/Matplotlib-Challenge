## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df=pd.merge(mouse_metadata,study_results,on="Mouse ID")

# Display the data table for preview
merged_df.head(15)
```

```python
# Checking the number of mice.
duplicates_mouse=merged_df["Mouse ID"].count()
print(duplicates_mouse)
without_duplicates=len(merged_df["Mouse ID"].unique())
print(without_duplicates)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_count = merged_df[merged_df.duplicated(subset=["Mouse ID","Timepoint"],keep="last")]
duplicate_mice_count.head(10)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data_df=merged_df.drop_duplicates(subset=["Mouse ID","Timepoint"],keep="last")
data_df.head(20)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function

tumor_summary_df=data_df.groupby("Drug Regimen")

tumor_mean=tumor_summary_df["Tumor Volume (mm3)"].mean()
tumor_med=tumor_summary_df["Tumor Volume (mm3)"].median()
tumor_var=tumor_summary_df["Tumor Volume (mm3)"].var()
tumor_std=tumor_summary_df["Tumor Volume (mm3)"].std()
tumor_sem=tumor_summary_df["Tumor Volume (mm3)"].sem()

stat_summary=pd.DataFrame({"Mean": tumor_mean,
                           "Median": tumor_med,
                           "Variance": tumor_var,
                           "Standard Deviation": tumor_std,
                           "Sem": tumor_sem})

stat_summary
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

mouse_df=data_df.loc[data_df["Mouse ID"]=="s185"]
cap_time=mouse_df["Timepoint"]
cap_volume=mouse_df["Tumor Volume (mm3)"]
plt.plot(cap_time,cap_volume,marker=".")

plt.title("Mouse s185: Timepoint vs Tumor Volume")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

capomulin_df=data_df.loc[data_df["Drug Regimen"]=="Capomulin"]

weight=capomulin_df["Weight (g)"]
tumor_vol=capomulin_df["Tumor Volume (mm3)"]
plt.scatter(weight,tumor_vol,marker=".")

plt.title("Mouse weight against Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient 
r_value=sts.pearsonr(weight,tumor_vol)
print(f"The corrlation/r-value={round(r_value[0],2)}")
```

```python
#linear regression model for mouse weight and average tumor volume for the Capomulin regimen
x=weight
y=tumor_vol

#Y=SLOPE*X+INTERCEPT
plt.scatter(weight,tumor_vol,marker=".")
plt.title("Mouse weight against Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

(slope,intercept,rvalue,pvalue,stderr)=sts.linregress(weight,tumor_vol)
tumor_vol_predicted=x * slope +intercept
regression= f"y ={round(slope,2)} * x + {round(intercept,2)}"


plt.plot(weight,tumor_vol_predicted,color="red")
plt.annotate(regression,(18,30),fontsize=12,color="red")

plt.show()
```
